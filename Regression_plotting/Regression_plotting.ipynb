```python
import matplotlib.pyplot as plt
import numpy as np

df.plot(kind = 'scatter', x = 'Season', y = 'Balance per club', figsize = (15, 6))

plt.title('Balance per club from 2004/05 - 2021/22',fontsize=21,fontfamily='serif',color='midnightblue')
plt.ylabel('Balance per club',fontsize=14)
plt.xlabel('Season',fontsize=14)
plt.annotate('only season with\nnegative balance\n per club in the\n last 16 seasons',fontsize=9,fontfamily='serif',color ='black',xy=(4, 20), xytext=(3, 3000000),
             arrowprops=dict(facecolor='red', shrink=0.01),
             )
plt.savefig('balance_per_club_04-22.png',dpi=500,bbox_inches = 'tight',facecolor='white')
```

```python
import matplotlib.pyplot as plt
import numpy as np

df.plot(kind = 'line', x = 'Season', y = 'Balance per club', figsize = (15, 6))

plt.title('Balance per club from 2004/05 - 2021/22',fontsize=21,fontfamily='serif',color='midnightblue')
plt.ylabel('Balance per club',fontsize=14)
plt.xlabel('Season',fontsize=14)
plt.annotate('only season with\nnegative balance\n per club in the\n last 16 seasons',fontsize=8,fontfamily='serif',color ='black',xy=(4, 20), xytext=(3.6, 3000000),
             arrowprops=dict(facecolor='red', shrink=0.01),
             )

plt.savefig('balance_per_club_04-22.png',dpi=500,bbox_inches = 'tight',facecolor='white')
```

```python
x = df['Starting year ']
y = df['Balance per club']

fit = np.polyfit(x, y, deg=1)

df.plot(kind='scatter', x='Starting year ', y='Balance per club', figsize=(15,6))

plt.title('Balance per club from 2004/05 - 2021/22',fontsize=21,fontfamily='serif',color='midnightblue')
plt.xlabel('Year of start of the season',fontsize=14)
plt.ylabel('Balance per club',fontsize=14)

plt.plot(x, fit[0] * x + fit[1], color='red')
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2015.5, 13500),fontsize=12,color='red')

plt.savefig('reg_lin_balance_per_club_04-22.png',dpi=500,bbox_inches = 'tight',facecolor='white')
```
